```python
import pandas as pd
import numpy as np

data=pd.read_json("last.json")

def odapuan(oda):
    if oda=="3 + 1":
        return 5
    elif oda=="2 + 1":
        return 3
    elif oda=="1 + 1":
        return 2
    else:
        return 0
def mkpuan(mk):
    liste=mk.split(" ")
    try:
        mks=int(liste[0])
        return (mks / 10)
    except:
        return 0
    
def yaspuan(yas):
    liste=yas.split(" ")
    yass= liste[0]
    if yass == "Sıfır":
        return 10
    else:
        return (8 / int(yass))
def acilmipuan(acil):
    if "acil" in acil.lower():
        return 5
    else: 
        return 0
    
pyaspuan=data["Yaş"].apply(yaspuan)
podapuan=data["Oda-Salon"].apply(odapuan)
pmkpuan=data["Metrekare"].apply(mkpuan)
pacilmipuan=data["İlan Başlığı"].apply(acilmipuan)

total=pyaspuan+podapuan+pmkpuan+pacilmipuan

data["Toplam Puan"]=total
tab=data.sort_values("Toplam Puan",ascending=False)

tab.to_excel("PuanlıEvler.xlsx")
```
